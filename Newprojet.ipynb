```python
import numpy as np
import pandas as pd

colonnes=["satisfaction_level","last_evaluation","number_project","average_montly_hours","time_spend_company",
         "Work_accident","Left","promotion_last_5years","Departement","Salary"]
x=pd.read_csv('HR_SBA_software.csv',header=1,names=colonnes)
x.info()
print(x)
```

```python
x = x.assign(Departement_sales=np.nan,Departement_accounting=np.nan,Departement_hr=np.nan,Departement_technical=np.nan
             ,Departement_support=np.nan,Departement_it=np.nan,
             Departement_marketing=np.nan,Departement_managment=np.nan,
             Departement_product_mng=np.nan,Departement_randD=np.nan)

x = x.assign(salary_low=np.nan,salary_medium=np.nan,salary_high=np.nan)
x.info()
```

```python
x.loc[(x['Departement'] == 'sales') , 'Departement_sales'] = 1
x.loc[(x['Departement'] != 'sales') , 'Departement_sales'] = 0
x.loc[(x['Departement'] == 'accounting') , 'Departement_accounting'] = 1
x.loc[(x['Departement'] != 'accounting') , 'Departement_accounting'] = 0
x.loc[(x['Departement'] == 'hr') , 'Departement_hr'] = 1
x.loc[(x['Departement'] != 'hr') , 'Departement_hr'] = 0
x.loc[(x['Departement'] == 'technical') , 'Departement_technical'] = 1
x.loc[(x['Departement'] != 'technical') , 'Departement_technical'] = 0
x.loc[(x['Departement'] == 'support') , 'Departement_support'] = 1
x.loc[(x['Departement'] != 'support') , 'Departement_support'] = 0
x.loc[(x['Departement'] == 'IT') , 'Departement_it'] = 1
x.loc[(x['Departement'] != 'IT') , 'Departement_it'] = 0
x.loc[(x['Departement'] == 'marketing') , 'Departement_marketing'] = 1
x.loc[(x['Departement'] != 'marketing') , 'Departement_marketing'] = 0
x.loc[(x['Departement'] == 'management') , 'Departement_managment'] = 1
x.loc[(x['Departement'] != 'management') , 'Departement_managment'] = 0
x.loc[(x['Departement'] == 'product_mng') , 'Departement_product_mng'] = 1
x.loc[(x['Departement'] != 'product_mng') , 'Departement_product_mng'] = 0
x.loc[(x['Departement'] == 'RandD') , 'Departement_randD'] = 1
x.loc[(x['Departement'] != 'RandD') , 'Departement_randD'] = 0
x.loc[(x['Salary'] == 'medium') , 'salary_medium'] = 1
x.loc[(x['Salary'] != 'medium') , 'salary_medium'] = 0
x.loc[(x['Salary'] == 'low') , 'salary_low'] = 1
x.loc[(x['Salary'] != 'low') , 'salary_low'] = 0
x.loc[(x['Salary'] == 'high') , 'salary_high'] = 1
x.loc[(x['Salary'] != 'high') , 'salary_high'] = 0
x.info()
```

```python
#Nous devons explicitement centrer et réduire les variables pour réaliser une ACP normée avec PCA.#
#Standardiser les données en utilisant la classe **StandardScaler** de la bibliothèque **sklearn**.
# Nous obtenons une matrice Z
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
model= sc.fit(x)
# Transformation des données selon le modèle construit
z= model.transform(x)
display(z)
```

```python
#Afficher la centrée réduite Z de StandardScaler() sous forme de matrice
columns=['last_evaluation','number_project','average_montly_hours','time_spend_company','Work_accident','Left','promotion_last_5years',
        'Departement_sales','Departement_accounting','Departement_hr','Departement_technical','Departement_support','Departement_it',
        'Departement_marketing','Departement_managment','Departement_product_mng','Departement_randD','salary_low','salary_medium',
        'salary_high']
        
z_df=pd.DataFrame(z,columns=columns)
print(z_df)
```

```python
## Vérifions les propriétés du nouvel ensemble de données : les moyennes et les écarts-type unitaires.
# Pour une matrice de données centrée et réduite, la moyenne de chaque colonne=0 et l'écarts_type de chaque colonne = 1
#verification pas numpy
import numpy as np
np.set_printoptions(suppress=True)
print(np.mean(z,axis=0))
print(np.std(z,axis=0))
```

```python
#Afficher et analyser la matrice de corrélation.
corr = np.corrcoef(z,rowvar=False) #rawvar c'est pour inverse la matrice
corr = pd.DataFrame(corr,columns=columns,index=columns)
corr
```

```python
# Importation classe PCA
from sklearn.decomposition import PCA
# instanciation d'un objet PCA
pca = PCA()

# nombre de composantes calculées
print(pca.n_components)
```

```python
# Transformation de Z (centrée réduite) en coordonnées factorielles (Y) selon le modèle PCA

# Y matrice des scores
y = pca.fit_transform(z)
```

```python
print(pca.n_components_)
```

```python
#Afficher Y
#Calculer et afficher les coordonnées factorielles (matrice des scores des individus)
columns = [' pca_%i ' % i for i in range(20)]
print(columns)
Y_df=pd.DataFrame(y,columns=columns)
print(Y_df)
```

```python
np.set_printoptions(suppress=True)
print(Y_df.corr())
print(np.corrcoef(Y_df,rowvar=False))
```

```python
# pour obtenir les variances (valeurs propres, λk) associées aux composantes principales.
val_propres= pca.explained_variance_
print(val_propres)
```

```python
# Les valeurs obtenues ne sont pas exactes.
# Règle: Somme des variances(Valeurs propres) = nombre de variables
print(np.sum(val_propres))
```

```python
# Il faut alors appliquer une correction.
# valeur corrigée
n=z_df.shape[0]
val_propres =(n -1)/n * pca.explained_variance_
print(val_propres)
```

```python
# Nous aurions pu obtenir les bonnes valeurs propres en passant par les valeurs singulières
#valeur propre
val_propres = pca.singular_values_ ** 2 / n
print(val_propres)
```

```python
#Tracer l’éboulis des valeurs propres (scree plot)
import matplotlib.pyplot as plt
plt.plot(range(pca.n_components_), val_propres )
plt.title('variance expliquée /CP')
plt.xlabel('Composantes principales')
plt.ylabel('Valeur de variance expliquée')
plt.xticks(range(pca.n_components_))
plt.show()
```

```python
plt.bar(range(pca.n_components_), pca.explained_variance_)
plt.title('Variance expliquée / CP')
plt.xlabel('Composantes principales')
plt.ylabel('Variance expliquée')
plt.xticks(range(pca.n_components_))
plt.show()

```

```python
# proportion de variance expliquée
#Calculer la proportion des variances (parts d’inertie) expliquées par chaque Composante  principale (CP)

print('proportion de variance expliquée')
print(pca.explained_variance_ratio_ ) # Il n’est pas nécessaire d’effectuer une correction dans ce cas

exp_var_ratio = pca.explained_variance_ratio_
print(exp_var_ratio*100)
#Somme cumulative des valeurs propres
cum_sum_eigenvalues = np.cumsum(exp_var_ratio)
print(cum_sum_eigenvalues)
```

```python
#Tracer le graphique des variances cumulées expliquées.
plt.plot(range(pca.n_components_),cum_sum_eigenvalues,'o-', linewidth=2,
color='blue')
plt.grid(which='both', linestyle='--')
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```
